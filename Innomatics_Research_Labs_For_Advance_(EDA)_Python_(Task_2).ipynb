#### **Q No. 1108 -**  ***[Defanging an IP Address](https://leetcode.com/problems/defanging-an-ip-address/description/)***

```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        return address.replace(".", "[.]")

# ---- User Input Section ----
raw_input_value = input()

# Remove "address =" part if user typed it
if "=" in raw_input_value:
    raw_input_value = raw_input_value.split("=", 1)[1].strip()

# Remove surrounding quotes
raw_input_value = raw_input_value.strip().replace('"', "").replace("'", "")

# Use Solution class
sol = Solution()
output = sol.defangIPaddr(raw_input_value)

# Print output with quotes
print(f'"{output}"')
```

#### **Q No. 1295 -** ***[Find Numbers with Even Number of Digit](https://leetcode.com/problems/find-numbers-with-even-number-of-digits/description/)***

```python
class Solution:
    def findNumbers(self, nums):
        count = 0
        for num in nums:
            if len(str(num)) % 2 == 0:
                count += 1
        return count

# ---- User Input ----
raw_input_value = input()

# Remove 'nums =' if present
if "=" in raw_input_value:
    raw_input_value = raw_input_value.split("=", 1)[1].strip()

# Convert string list to actual list
nums = eval(raw_input_value)

# Use Solution class
sol = Solution()
print(sol.findNumbers(nums))
```

#### **Q No. 1512 -** ***[Numbers of Good Pairs](https://leetcode.com/problems/number-of-good-pairs/description/)***

```python
class Solution:
    def numIdenticalPairs(self, nums):
        freq = {}
        count = 0
        for num in nums:
            if num in freq:
                count += freq[num]
                freq[num] += 1
            else:
                freq[num] = 1
        return count

# ---- User Input ----
raw_input_value = input()

# Remove 'nums =' if present
if "=" in raw_input_value:
    raw_input_value = raw_input_value.split("=", 1)[1].strip()

nums = eval(raw_input_value)

# Run solution
sol = Solution()
print(sol.numIdenticalPairs(nums))
```

#### **Q No. 1365 -** ***[How Many Numbers Are Small Than the Current Number](https://leetcode.com/problems/how-many-numbers-are-smaller-than-the-current-number/description/)***

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums):
        freq = [0] * 101

        for num in nums:
            freq[num] += 1

        for i in range(1, 101):
            freq[i] += freq[i - 1]

        result = []
        for num in nums:
            result.append(freq[num - 1] if num > 0 else 0)

        return result

# ---- User Input ----
raw_input_value = input()

# Remove 'nums =' if present
if "=" in raw_input_value:
    raw_input_value = raw_input_value.split("=", 1)[1].strip()

nums = eval(raw_input_value)

# Run solution
sol = Solution()
print(sol.smallerNumbersThanCurrent(nums))
```

#### **Q No. 1281 -** ***[Subtract the Product and Sum of Digits of an Integer](https://leetcode.com/problems/subtract-the-product-and-sum-of-digits-of-an-integer/description/)***

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        product = 1
        total = 0

        while n > 0:
            digit = n % 10
            product *= digit
            total += digit
            n //= 10

        return product - total

# ---- User Input ----
raw_input_value = input()

# Remove 'n =' if present
if "=" in raw_input_value:
    raw_input_value = raw_input_value.split("=", 1)[1].strip()

n = int(raw_input_value)

# Run solution
sol = Solution()
print(sol.subtractProductAndSum(n))
```

#### **Q No. 1486 -** ***[XOR Operations in a Array](https://leetcode.com/problems/xor-operation-in-an-array/description/)***

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result = 0
        for i in range(n):
            result ^= start + 2 * i
        return result

# ---- User Input ----
raw_input = input()

# Remove spaces
raw_input = raw_input.replace(" ", "")

# Extract n and start
parts = raw_input.split(",")

n = int(parts[0].split("=")[1])
start = int(parts[1].split("=")[1])

# Run solution
sol = Solution()
print(sol.xorOperation(n, start))
```
