#### **Q 1464 -**  ***[Maximum Product of Two Elements in an Array](https://leetcode.com/problems/maximum-product-of-two-elements-in-an-array/description/)***

```python
class Solution:
    def maxProduct(self, nums):
        nums.sort()
        return (nums[-1] - 1) * (nums[-2] - 1)

# ---- User Input ----
raw_input_value = input()

# Remove 'nums =' if present
if "=" in raw_input_value:
    raw_input_value = raw_input_value.split("=", 1)[1].strip()

nums = eval(raw_input_value)

# Run solution
sol = Solution()
print(sol.maxProduct(nums))
```

#### **Q 1395 -**  ***[Count Number of Teams](https://leetcode.com/problems/count-number-of-teams/description/)***

```python
class Solution:
    def numTeams(self, rating):
        n = len(rating)
        teams = 0

        for j in range(n):
            left_smaller = left_greater = 0
            right_smaller = right_greater = 0

            for i in range(j):
                if rating[i] < rating[j]:
                    left_smaller += 1
                else:
                    left_greater += 1

            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    right_greater += 1
                else:
                    right_smaller += 1

            teams += left_smaller * right_greater
            teams += left_greater * right_smaller

        return teams

# ---- User Input ----
raw_input_value = input()

# Remove 'rating =' if present
if "=" in raw_input_value:
    raw_input_value = raw_input_value.split("=", 1)[1].strip()

rating = eval(raw_input_value)

# Run solution
sol = Solution()
print(sol.numTeams(rating))
```

#### **Q 1450 -**  ***[Number of Students Doing Homework at a Given Time](https://leetcode.com/problems/number-of-students-doing-homework-at-a-given-time/description/)***

```python
class Solution:
    def busyStudent(self, startTime, endTime, queryTime):
        count = 0
        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1
        return count

# ---- User Input ----
raw_input = input()

# Remove spaces
raw_input = raw_input.replace(" ", "")

# Parse input string to extract startTime, endTime, and queryTime
# Example: raw_input = "startTime=[1,2,3],endTime=[3,2,7],queryTime=4"

# Extract startTime list string
start_idx_st = raw_input.find("startTime=") + len("startTime=")
end_idx_st = raw_input.find("]", start_idx_st) + 1
startTime_str = raw_input[start_idx_st : end_idx_st]

# Extract endTime list string
start_idx_et = raw_input.find("endTime=") + len("endTime=")
end_idx_et = raw_input.find("]", start_idx_et) + 1
endTime_str = raw_input[start_idx_et : end_idx_et]

# Extract queryTime integer string
queryTime_key_idx = raw_input.find("queryTime=")
queryTime_str = raw_input[queryTime_key_idx + len("queryTime="):]

startTime = eval(startTime_str)
endTime = eval(endTime_str)
queryTime = int(queryTime_str)

# Run solution
sol = Solution()
print(sol.busyStudent(startTime, endTime, queryTime))
```

#### **Q 1342 -**  ***[Number of Steps to Reduce a Number to Zero](https://leetcode.com/problems/number-of-steps-to-reduce-a-number-to-zero/description/)***

```python
class Solution:
    def numberOfSteps(self, num: int) -> int:
        steps = 0

        while num > 0:
            if num % 2 == 0:
                num //= 2
            else:
                num -= 1
            steps += 1

        return steps

# ---- User Input ----
raw_input_value = input()

# Remove 'num =' if present
if "=" in raw_input_value:
    raw_input_value = raw_input_value.split("=", 1)[1].strip()

num = int(raw_input_value)

# Run solution
sol = Solution()
print(sol.numberOfSteps(num))
```

#### **Q 338 -**  ***[Counting Bits](https://leetcode.com/problems/counting-bits/description/)***

```python
class Solution:
    def countBits(self, n: int):
        ans = [0] * (n + 1)

        for i in range(1, n + 1):
            ans[i] = ans[i // 2] + (i % 2)

        return ans

# ---- User Input ----
raw_input_value = input()

# Remove 'n =' if present
if "=" in raw_input_value:
    raw_input_value = raw_input_value.split("=", 1)[1].strip()

n = int(raw_input_value)

# Run solution
sol = Solution()
print(sol.countBits(n))
```
